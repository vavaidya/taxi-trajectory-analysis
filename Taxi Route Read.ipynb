```python
import pandas as pd
import numpy as np
```

```python
chunks = pd.read_csv("taxi_route.csv", chunksize=1000000)
```

```python
chunk_list = []  # append each chunk df here 

# Each chunk is in df format
for chunk in chunks:  
    # Once the data filtering is done, append the chunk to list
    chunk_list.append(chunk)
    
routes = pd.concat(chunk_list, axis= 0)
```

```python
routes
```

```python
import ast

def convert_string(string):
    array = np.array(ast.literal_eval(string))
    return np.flip(array)
```

```python
%%time
routes['polyline array'] = routes['POLYLINE'].map(convert_string)
```

```python
import pickle
pickle.dump( routes, open( "save.p", "wb" ) )
```
